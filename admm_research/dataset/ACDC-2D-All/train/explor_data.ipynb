```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
from pathlib import Path   
from typing import *
import re
```

```python
RV_list: List[Path] = list(Path('admm_research/dataset/ACDC-2D-All/train/RV_prior').rglob('*.png'))
print(RV_list.__len__())
print(RV_list[:5])
r_pattern = re.compile(r'patient\d+_\d+')
unique_patient = sorted(list(set([r_pattern.findall(x.stem)[0] for x in RV_list])))
print(unique_pateint)
```

```python
def group_patients(all_path:List[Path], r_unique_list:List[str])->Dict[str,List[Path]]:
    assert isinstance(all_path,list)
    assert isinstance(all_path[0],Path)
    assert isinstance(r_unique_list,list)
    assert isinstance(r_unique_list[0],str)
    for l in all_path:
        assert l.exists(), l
    mapping:Dict[str,List[Path]] = {}
    for r_unique in r_unique_list:
        _r_unique = re.compile(r_unique)
        sublist = sorted([x for x in all_path if len(_r_unique.findall(x.stem))>0 is not None])
        mapping[r_unique]= sublist
    return mapping 

def read_numpy_tensor(unique_patient:str, mapping:dict)->np.ndarray:
    assert isinstance(unique_patient,str)
    assert isinstance(mapping,dict)
    # try:
    volumns = np.stack(
        [np.array(Image.open(x)) for x in mapping.get(unique_patient)]
    )
    # except AttributeError:
    #     raise AttributeError(f'Given key words:{unique_pateint} not founded in mapping with keys: {mapping.keys()}')
    
    return volumns
    
```

```python
mapping = group_patients(RV_list,unique_patient)
print(mapping)
```

```python
max_value_dict = {}
for p in unique_patient:
    print(p)
    patient = read_numpy_tensor(unique_patient=p,mapping=mapping)
    print(patient.shape)
    max_value = patient.max()
    max_value_dict[p]=max_value
    
```
